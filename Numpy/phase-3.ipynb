```python
import numpy as np
import matplotlib.pyplot as plt

# import matplot.pyplot as plt

sales_data = np.array([
    [1, 1000, 2000, 3000, 4000 ],
    [2, 5000, 6000, 7000, 8000],
    [3, 9000, 10000, 11000, 12000],
    [4, 13000, 14000, 15000, 16000],
    [5, 17000, 18000, 19000, 20000]
])


print("========= zomato sales analysis =========")
print("\n sales data shapes, ", sales_data.shape)
print("\n sample data for 1st 3 rest ",sales_data[:3])


# total sales per year

print(np.sum(sales_data, axis=0))
yeary_total = np.sum(sales_data[:,:1:],axis=0)
print(yeary_total)


#  minimum sales for per restaurant
#  here axis 1 is using for row, axis = 0 for column
min_sales = np.min(sales_data[:,1:],axis=1)
print(min_sales)


# amx sale

max_sales=np.max(sales_data[:,1:],axis=0)

# avg sales per restaurant
avg_sales = np.mean(sales_data[:,1:],axis = 1)

# learn all math ops


# c,umsum
cumsum= np.cumsum(sales_data[:,1:],axis = 1)
print("here we are cusum",cumsum)



plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("avg sales of all sales")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()


# VECTOR

vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print("vector addition" , vector1 + vector2)

print("\n multiply", vector1 * vector2)

print("\n dot product", np.dot(vector1 , vector2))



# angle of a vecor
angle = np.arccos(np.dot(vector1,vector2)/ (np.linalg.norm(vector1)* np.linalg.norm(vector2)))
print("angle",angle)




# vectorize
rest_type = np.array(['pure veg', 'pizza', 'burger','biryani'])
vectorize_upper = np.vectorize(str.upper)
print("vectorize upper str",vectorize_upper(rest_type))


#  broadcasting in vectors
montly_avg = sales_data[:,1:] / 12
print("montly avg", montly_avg)
```
